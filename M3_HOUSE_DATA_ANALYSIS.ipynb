```python
#DATA IMPORTATION#
##import files##
from google.colab import files
##import pandas library##
import pandas as pd
##import the input and output necessities##
import io
```

```python
##upload the csv file from the PC##
upload=files.upload()
```

```python
##print the upload keys from the dictionary##
upload.keys()
```

```python
##read the csv file as a dataframe##
file_name=next(iter(upload))
assign=pd.read_csv(io.BytesIO(upload[file_name]))
assign
```

```python
##reading the head and tail to confirm our data##
##head##
assign.head(10)
```

```python
##tail##
assign.tail(10)
```

```python
#DATA CLEANING#
##checking for duplictate values##
assign.duplicated().sum()
```

```python
##checking for missing values##
assign.isna().sum()
```
